## Calculating the metrics

```
import numpy as np
import pandas as pd
import os

# Define datasets, seeds, percentages, and combinations
datasets = ['algo004', 'comp', 'ml', 'virtualshakespeare']
seeds = [18, 61, 53, 29, 69, 42, 2, 21, 78, 99]
percentages = [0.05, 0.1, 0.15, 0.2, 0.25, 0.5, 0.75]
combinations = [
    ['algo004', 'comp'],
    ['algo004', 'ml'],
    ['algo004', 'virtualshakespeare'],
    ['comp', 'ml'],
    ['comp', 'virtualshakespeare'],
    ['ml', 'virtualshakespeare'],
    ['algo004', 'comp', 'ml'],
    ['algo004', 'comp', 'virtualshakespeare'],
    ['algo004', 'ml', 'virtualshakespeare'],
    ['comp', 'ml', 'virtualshakespeare'],
    ['algo004', 'comp', 'ml', 'virtualshakespeare']
]

# Function to calculate metrics from confusion matrix
def calculate_metrics(cm):
    tn, fp, fn, tp = cm.ravel()
    total_samples = tn + fp + fn + tp
    # Precision
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    # True Positives divided by all samples
    tp_div_all = tp / total_samples
    # False Positives divided by all samples
    fp_div_all = fp / total_samples
    return precision, tp_div_all, fp_div_all

# Main processing
for seed in seeds:
    for percentage in percentages:
        percentage_str = f'{percentage}p'  # Convert to string with 'p' suffix
        for ds in datasets:
            # Initialize a list to collect results for this dataset, seed, and percentage
            results = []
            # First, collect the individual dataset's metrics
            base_dir = f'/content/drive/MyDrive/Colab Notebooks/Fed Learning Research/{ds}_results'
            cm_npy_dir = os.path.join(base_dir, 'cm_npy', str(seed))
            cm_filename = f'cm_{ds}_{percentage_str}_seed{seed}.npy'
            cm_path = os.path.join(cm_npy_dir, cm_filename)
            if os.path.exists(cm_path):
                cm = np.load(cm_path)
                precision, tp_div_all, fp_div_all = calculate_metrics(cm)
                result = {
                    'Dataset': ds,
                    'Combination': ds,
                    'Percentage': percentage,
                    'Seed': seed,
                    'Precision': precision,
                    'TP_div_all': tp_div_all,
                    'FP_div_all': fp_div_all
                }
                results.append(result)
            else:
                print(f'Confusion matrix not found for {ds} at {percentage}% seed {seed}')

            # Now, collect metrics for combinations that include this dataset
            for combo in combinations:
                if ds in combo:
                    combo_name = '_'.join(combo)
                    cm_npy_dir = f'/content/drive/MyDrive/Colab Notebooks/Fed Learning Research/combined_results/cm_npy/npy_{percentage}/{seed}'
                    cm_filename = f'cm_{ds}_{combo_name}_{percentage_str}_seed{seed}.npy'
                    cm_path = os.path.join(cm_npy_dir, cm_filename)
                    if os.path.exists(cm_path):
                        cm = np.load(cm_path)
                        precision, tp_div_all, fp_div_all = calculate_metrics(cm)
                        result = {
                            'Dataset': ds,
                            'Combination': combo_name,
                            'Percentage': percentage,
                            'Seed': seed,
                            'Precision': precision,
                            'TP_div_all': tp_div_all,
                            'FP_div_all': fp_div_all
                        }
                        results.append(result)
                    else:
                        print(f'Confusion matrix not found for {ds} in combination {combo_name} at {percentage}% seed {seed}')

            # Save the collected results into a CSV file
            if results:
                results_df = pd.DataFrame(results)
                # Define the save path
                csv_dir = f'/content/drive/MyDrive/Colab Notebooks/Fed Learning Research/{ds}_results/csv/combined_results'
                os.makedirs(csv_dir, exist_ok=True)
                csv_filename = f'metrics_{ds}_seed{seed}_{percentage_str}.csv'
                csv_save_path = os.path.join(csv_dir, csv_filename)
                results_df.to_csv(csv_save_path, index=False)
            else:
                print(f'No results to save for {ds} at {percentage}% seed {seed}')
```

```
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define datasets, seeds, percentages
datasets = ['algo004', 'comp', 'ml', 'virtualshakespeare']
seeds = [18, 61, 53, 29, 69, 42, 2, 21, 78, 99]
percentages = [0.05, 0.1, 0.15, 0.2, 0.25, 0.5, 0.75]  # percentages as integers

# Define difference metrics
difference_metrics = ['Precision_Difference', 'TP_div_all_Difference', 'FP_div_all_Difference']

# Define colors
positive_color = 'blue'
negative_color = 'red'

# Parent directory for charts
charts_parent_dir = '/content/drive/MyDrive/Colab Notebooks/GNN Research/charts'

# Main processing
for ds in datasets:
    for seed in seeds:
        for percentage in percentages:
            # Construct percentage string
            percentage_str = f'{percentage}p'
            # Path to the CSV file
            csv_file = f'/content/drive/MyDrive/Colab Notebooks/GNN Research/{ds}_results/csv/combined_csv/metrics_{ds}_seed{seed}_{percentage_str}.csv'
            # Check if the CSV file exists
            if not os.path.exists(csv_file):
                print(f'CSV file not found: {csv_file}')
                continue
            # Read the CSV file
            df = pd.read_csv(csv_file)
            # Skip if there are no combinations (less than 2 rows)
            if df.shape[0] <= 1:
                print(f'Not enough data to plot for {ds}, seed {seed}, percentage {percentage}')
                continue
            # For each difference metric, create a chart
            for metric in difference_metrics:
                # Prepare data
                combinations = df['Combination'][1:]  # Exclude the first row (individual dataset)
                differences = df[metric][1:]  # Exclude the first row (individual dataset)
                # Convert combinations and differences to lists
                combinations = combinations.tolist()
                differences = differences.tolist()
                # Create a color list based on positive or negative differences
                colors = [positive_color if x >= 0 else negative_color for x in differences]
                # Create the plot
                plt.figure(figsize=(10, len(combinations)*0.5))
                # Plot horizontal bars
                plt.barh(combinations, differences, color=colors)
                # Add vertical line at x=0
                plt.axvline(x=0, color='black', linewidth=0.8)
                # Set labels and title
                plt.xlabel(metric.replace('_', ' '))
                plt.ylabel('Combination')
                plt.title(f'{metric.replace("_", " ")} for {ds} (Seed {seed}, {percentage}%)')
                plt.tight_layout()
                # Define the save path
                chart_dir = os.path.join(charts_parent_dir, ds, f'seed_{seed}', f'{percentage}p')
                os.makedirs(chart_dir, exist_ok=True)
                # Name the file as per your specification
                metric_name = metric.replace('_Difference', '').replace('_', '')
                for idx, combination in enumerate(combinations):
                    # Prepare data for single combination
                    diff = differences[idx]
                    color = colors[idx]
                    # Create a single-bar plot
                    plt.figure(figsize=(6, 1))
                    plt.barh([combination], [diff], color=color)
                    plt.axvline(x=0, color='black', linewidth=0.8)
                    plt.xlabel(metric_name)
                    plt.title(f'{metric_name} Difference for {ds} with {combination}\n(Seed {seed}, {percentage}%)')
                    plt.tight_layout()
                    # Filename: dataset_combination_percent_seed_metric.png
                    chart_filename = f'{ds}_{combination}_{percentage}p_seed{seed}_{metric_name}.png'
                    chart_save_path = os.path.join(chart_dir, chart_filename)
                    # Save the plot
                    plt.savefig(chart_save_path, dpi=200)
                    plt.close()
                    #print(f'Saved chart: {chart_save_path}')
                # Comment out the group plot if you prefer individual charts per combination
                plt.savefig(chart_save_path, dpi=200)
                plt.close()
                print(f'Saved chart: {chart_save_path}')
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.ticker import MaxNLocator

#################
# Configuration #
#################

datasets = ['algo004', 'comp', 'ml', 'virtualshakespeare']
percentages = [0.05, 0.1, 0.15, 0.2, 0.25, 0.5, 0.75]

metrics_info = {
    'Mean_AUC': {'label': 'Mean AUC'},
    'Mean_Accuracy': {'label': 'Mean Accuracy'},
    'Mean_Precision': {'label': 'Mean Precision'},
}

pretty_names = {
    'algo004': 'Algo',
    'comp': 'Comp',
    'ml': 'ML',
    'virtualshakespeare': 'Shake'
}

mean_results_dir = '/content/drive/MyDrive/Colab Notebooks/GNN Research/mean_results'
charts_dir = '/content/drive/MyDrive/Colab Notebooks/GNN Research/mean_charts'

# Example pairwise CSV
pairwise_csv = "/content/drive/MyDrive/Colab Notebooks/GNN Research/Graph Neural Network/degree_dist_comparison.csv"

# Adjust or add these for triplets/quads if desired
triplet_csv = "/content/drive/MyDrive/Colab Notebooks/GNN Research/Graph Neural Network/triplet_wasserstein.csv"
quad_csv = "/content/drive/MyDrive/Colab Notebooks/GNN Research/Graph Neural Network/quadruplet_wasserstein.csv"

######################################
# 1) Read Pairwise CSV
######################################
df_pair = pd.read_csv(pairwise_csv)
wasserstein_map = {}
for row in df_pair.itertuples(index=False):
    ds1 = row.Dataset1.strip().lower()
    ds2 = row.Dataset2.strip().lower()
    pct = row.Percent
    w_value = row.Mean_Wasserstein_Degree
    sorted_pair = tuple(sorted([ds1, ds2]))
    wasserstein_map[(sorted_pair, pct)] = w_value

######################################
# 2) Read Triplet CSV
######################################
df_trip = pd.read_csv(triplet_csv)
triplet_w_map = {}
for row in df_trip.itertuples(index=False):
    ds_list = [x.strip().lower() for x in row.Datasets.split(',')]
    ds_tuple = tuple(sorted(ds_list))
    pct = row.Percent
    w_val = row.Triplet_Wasserstein
    triplet_w_map[(ds_tuple, pct)] = w_val

######################################
# 3) Read Quadruplet CSV
######################################
df_quad = pd.read_csv(quad_csv)
quad_w_map = {}
for row in df_quad.itertuples(index=False):
    ds_list = [x.strip().lower() for x in row.Datasets.split(',')]
    ds_tuple = tuple(sorted(ds_list))
    pct = row.Percent
    w_val = row.Quad_Wasserstein
    quad_w_map[(ds_tuple, pct)] = w_val

os.makedirs(charts_dir, exist_ok=True)

#######################
# 4) Main Plot Logic
#######################
for ds in datasets:
    for percentage in percentages:
        pct_str = f'{percentage}p'

        csv_file = os.path.join(mean_results_dir, ds, f'mean_metrics_{ds}_{pct_str}.csv')
        if not os.path.exists(csv_file):
            print(f'CSV not found: {csv_file}')
            continue

        df = pd.read_csv(csv_file)
        if df.empty:
            print(f'No data to plot for {ds} at {percentage}%')
            continue

        plot_save_dir = os.path.join(charts_dir, ds, pct_str)
        os.makedirs(plot_save_dir, exist_ok=True)

        for metric, info in metrics_info.items():
            y_label = info['label']

            baseline_rows = df[df['Combination'].str.lower() == ds.lower()]
            if baseline_rows.empty:
                print(f"No baseline row for {ds} in {csv_file}. Skipping {metric}.")
                continue

            baseline_value = baseline_rows[metric].values[0]
            df['Difference'] = df[metric] - baseline_value

            # Build y-labels
            y_labels = []
            for combo_str in df['Combination']:
                parts = [x.strip().lower() for x in combo_str.split('_')]
                pretty_parts = [pretty_names.get(p, p) for p in parts]
                y_labels.append(" + ".join(pretty_parts))

            differences = df['Difference'].values
            bar_colors = ['#0000CC' if d >= 0 else '#8B0000' for d in differences]

            fig, ax = plt.subplots(figsize=(10, 6))
            y_positions = np.arange(len(df))
            ax.barh(y_positions, differences, color=bar_colors, align='center')

            # Basic styling
            ax.set_axisbelow(True)
            ax.grid(True, axis='x', linestyle='--', alpha=0.5)
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)

            ax.set_yticks(y_positions)
            ax.set_yticklabels(y_labels, fontsize=11)
            ax.xaxis.set_major_locator(MaxNLocator(6))

            # Title: multiply percentage by 100 (0.05 => 5%)
            ax.set_title(f"{y_label} comparison for {pretty_names[ds]} at {percentage * 100:.0f}%", fontsize=15)
            ax.set_xlabel(f"Difference in {y_label} (vs. baseline)", fontsize=13)

            # Increase left margin to avoid collisions
            # e.g. 0.25 or 0.28 if you need more
            plt.subplots_adjust(left=0.25, right=0.90)

            # Annotate bars: Metric above, W below
            for i, row_data in df.iterrows():
                combo_str = row_data['Combination']
                diff_val = row_data['Difference']
                actual_val = row_data[metric]

                if pd.notnull(actual_val):
                    metric_text = f"{y_label.split()[-1]}: {actual_val:.3f}"
                else:
                    metric_text = f"{y_label.split()[-1]}: NaN"

                offset = 0.0001 * abs(diff_val if diff_val else 1)
                if diff_val >= 0:
                    base_x = diff_val + offset
                    ha_align = 'left'
                else:
                    base_x = diff_val - offset
                    ha_align = 'right'

                # Metric above
                ax.text(base_x, i + 0.15, metric_text,
                        va='center',
                        ha=ha_align,
                        fontsize=9,
                        color='black')

                # Identify pair vs triplet vs quad
                parts_underscore = combo_str.split('_')
                ds_list_sorted = tuple(sorted(p.strip().lower() for p in parts_underscore))
                w_line = None

                if len(ds_list_sorted) == 2:
                    # Pair
                    if (ds_list_sorted, percentage) in wasserstein_map:
                        w_val = wasserstein_map[(ds_list_sorted, percentage)]
                        w_line = f"W={w_val:.3f}"
                elif len(ds_list_sorted) == 3:
                    # Triplet
                    if (ds_list_sorted, percentage) in triplet_w_map:
                        w_val = triplet_w_map[(ds_list_sorted, percentage)]
                        w_line = f"W3={w_val:.3f}"
                elif len(ds_list_sorted) == 4:
                    # Quad
                    if (ds_list_sorted, percentage) in quad_w_map:
                        w_val = quad_w_map[(ds_list_sorted, percentage)]
                        w_line = f"W4={w_val:.3f}"

                # If we have a W value, place it below
                if w_line is not None:
                    ax.text(base_x, i - 0.10, w_line,
                            va='center',
                            ha=ha_align,
                            fontsize=9,
                            color='#00994C')

            outname = f'{ds}_{int(percentage*100)}pct_{metric}_horizontal_diff.png'
            outpath = os.path.join(plot_save_dir, outname)
            plt.savefig(outpath, dpi=150)
            plt.close()
            print(f"Saved plot: {outpath}")
```

## Hypothesis Testing

```
import os
import numpy as np
import pandas as pd
from scipy import stats

# Define datasets, percentages, and combinations
datasets = ['algo004', 'comp', 'ml', 'virtualshakespeare']
percentages = [0.05, 0.1, 0.15, 0.2, 0.25, 0.5, 0.75]  # percentages as integers
num_seeds = 10  # Number of seeds used in your experiments

# Significance levels
alphas = [0.1, 0.05, 0.01]

# Directories
base_dir = '/content/drive/MyDrive/Colab Notebooks/GNN Research'
mean_results_dir = os.path.join(base_dir, 'mean_results')
hypothesis_testing_dir = os.path.join(base_dir, 'hypothesis_testing')
os.makedirs(hypothesis_testing_dir, exist_ok=True)

# Perform hypothesis testing
for ds in datasets:
    # Create a directory for the dataset
    ds_hypothesis_dir = os.path.join(hypothesis_testing_dir, ds)
    os.makedirs(ds_hypothesis_dir, exist_ok=True)

    for percentage in percentages:
        percentage_str = f'{percentage}p'

        # Path to the mean metrics CSV for this dataset and percentage
        mean_csv_path = os.path.join(mean_results_dir, ds, f'mean_metrics_{ds}_{percentage_str}.csv')

        if not os.path.exists(mean_csv_path):
            print(f'Mean metrics CSV not found for {ds} at {percentage}%')
            continue

        # Load the mean metrics
        mean_df = pd.read_csv(mean_csv_path)

        # Extract the individual dataset row (base case)
        individual_row = mean_df[mean_df['Combination'] == ds]
        if individual_row.empty:
            print(f'No individual data found for {ds} at {percentage}%')
            continue

        # Get individual dataset statistics
        ind_mean_auc = individual_row['Mean_AUC'].values[0]
        ind_std_auc = individual_row['Std_AUC'].values[0]
        ind_mean_precision = individual_row['Mean_Precision'].values[0]
        ind_std_precision = individual_row['Std_Precision'].values[0]

        # Filter out the combinations (excluding the individual dataset itself)
        combination_rows = mean_df[mean_df['Combination'] != ds]

        # Initialize a list to store hypothesis test results
        hypothesis_results = []

        for _, row in combination_rows.iterrows():
            combination_name = row['Combination']

            # Get combination statistics
            comb_mean_auc = row['Mean_AUC']
            comb_std_auc = row['Std_AUC']
            comb_mean_precision = row['Mean_Precision']
            comb_std_precision = row['Std_Precision']

            # Check for valid data
            if pd.isna(comb_mean_auc) or pd.isna(comb_std_auc) or pd.isna(comb_mean_precision) or pd.isna(comb_std_precision):
                print(f'Skipping combination {combination_name} due to NaN values at {percentage}%')
                continue

            # Perform one-sided t-tests as per your specified hypotheses
            # Null Hypothesis: Mean individual <= Mean combined
            # Alternative Hypothesis: Mean individual > Mean combined

            # AUC t-test
            auc_t_stat, auc_p_value = stats.ttest_ind_from_stats(
                mean1=ind_mean_auc, std1=ind_std_auc, nobs1=num_seeds,
                mean2=comb_mean_auc, std2=comb_std_auc, nobs2=num_seeds,
                equal_var=False, alternative='greater'
            )

            # Precision t-test
            precision_t_stat, precision_p_value = stats.ttest_ind_from_stats(
                mean1=ind_mean_precision, std1=ind_std_precision, nobs1=num_seeds,
                mean2=comb_mean_precision, std2=comb_std_precision, nobs2=num_seeds,
                equal_var=False, alternative='greater'
            )

            # Store results for each alpha level
            for alpha in alphas:
                # Determine decision for AUC
                auc_decision = 'Reject Null Hypothesis' if auc_p_value < alpha else 'Fail to Reject Null Hypothesis'

                # Determine decision for Precision
                precision_decision = 'Reject Null Hypothesis' if precision_p_value < alpha else 'Fail to Reject Null Hypothesis'

                # Store results
                result = {
                    'Dataset': ds,
                    'Percentage': percentage,
                    'Combination': combination_name,
                    'Alpha': alpha,
                    'AUC_T-Statistic': auc_t_stat,
                    'AUC_P-Value': auc_p_value,
                    'AUC_Decision': auc_decision,
                    'Precision_T-Statistic': precision_t_stat,
                    'Precision_P-Value': precision_p_value,
                    'Precision_Decision': precision_decision
                }
                hypothesis_results.append(result)

        # Save the hypothesis test results to a CSV file
        if hypothesis_results:
            hypothesis_df = pd.DataFrame(hypothesis_results)
            hypothesis_save_path = os.path.join(ds_hypothesis_dir, f'hypothesis_testing_{ds}_{percentage_str}.csv')
            hypothesis_df.to_csv(hypothesis_save_path, index=False)
            print(f'Hypothesis test results saved for {ds} at {percentage}%: {hypothesis_save_path}')
        else:
            print(f'No hypothesis test results for {ds} at {percentage}%')
```

# Hypothesis Testing for CNN and GNN

```
import os
import numpy as np
import pandas as pd
from scipy import stats

# Configuration
datasets    = ['algo004', 'comp', 'ml', 'virtualshakespeare']
seeds       = [18, 61, 53, 29, 69, 42, 2, 21, 78, 99]
percentages = [0.10, 0.25, 0.50, 0.75]

cnn_results_dir = '/content/drive/MyDrive/Colab Notebooks/GNN Research/Convolutional NN'
gnn_mean_dir    = '/content/drive/MyDrive/Colab Notebooks/GNN Research/mean_results'
t_test_dir      = '/content/drive/MyDrive/Colab Notebooks/GNN Research/mean_results_t_tests'
os.makedirs(t_test_dir, exist_ok=True)

alpha      = 0.1
full_combo = '_'.join(datasets)  # "algo004_comp_ml_virtualshakespeare"

results = []

for ds in datasets:
    for perc in percentages:
        # 1) gather CNN AUCs
        cnn_aucs = []
        for seed in seeds:
            path = os.path.join(
                cnn_results_dir, ds, 'csv', str(seed),
                f'results_{ds}_seed{seed}.csv'
            )
            if os.path.exists(path):
                df = pd.read_csv(path)
                cnn_aucs.append(df.loc[0, 'AUC'])
        if not cnn_aucs:
            continue
        cnn_mean = np.mean(cnn_aucs)
        cnn_std  = np.std(cnn_aucs, ddof=1)

        # 2) load GNN combined mean AUC & std
        mean_path = os.path.join(
            gnn_mean_dir, ds,
            f'mean_metrics_{ds}_{perc}p.csv'
        )
        if not os.path.exists(mean_path):
            continue
        mean_df = pd.read_csv(mean_path)
        row = mean_df[mean_df['Combination'] == full_combo]
        if row.empty:
            continue
        gnn_mean = row['Mean_AUC'].iloc[0]
        gnn_std  = row['Std_AUC'].iloc[0]

        # 3) one-sided Welch's t-test (H0: CNN_mean >= GNN_mean vs H1: CNN_mean < GNN_mean)
        t_stat, p_value = stats.ttest_ind_from_stats(
            mean1=cnn_mean, std1=cnn_std, nobs1=len(cnn_aucs),
            mean2=gnn_mean, std2=gnn_std, nobs2=len(seeds),
            equal_var=False, alternative='less'
        )

        # 4) decision at alpha=0.1
        decision = 'Reject H0' if p_value < alpha else 'Fail to Reject H0'

        # 5) append to results
        results.append({
            'Dataset': ds,
            'Percentage': perc,
            'CNN_Mean_AUC': cnn_mean,
            'CNN_Std_AUC': cnn_std,
            'GNN_Mean_AUC': gnn_mean,
            'GNN_Std_AUC': gnn_std,
            'T-Statistic': t_stat,
            'P-Value': p_value,
            f'Decision (alpha={alpha})': decision
        })

# Save aggregated t-test results to CSV
results_df = pd.DataFrame(results)
save_path = os.path.join(t_test_dir, 't_test_mean_results.csv')
results_df.to_csv(save_path, index=False)
print(f"Saved combined t-test results (alpha={alpha}) to {save_path}")
```

```
import os
import pandas as pd
import numpy as np

# Configuration
datasets = ["algo004", "ml", "virtualshakespeare", "comp"]
seeds = ["18", "61", "53", "29", "69", "42", "2", "21", "78", "99"]
# Map the folder names (10,25,50,75) to decimal percentages
percent_folders = ["10", "25", "50", "75"]
percent_map = {"10": 0.10, "25": 0.25, "50": 0.50, "75": 0.75}

# Base paths
base_conv_dir = "drive/MyDrive/Colab Notebooks/GNN Research/Convolutional NN"
base_mean_dir = "drive/MyDrive/Colab Notebooks/GNN Research/mean_results"

# Ensure the mean_results folder (and its subfolders) exist
for ds in datasets:
    out_dir = os.path.join(base_mean_dir, ds)
    os.makedirs(out_dir, exist_ok=True)

# Loop over each dataset
for ds in datasets:
    # Prepare a list to collect per-percent mean values
    summary = []

    for pct_folder in percent_folders:
        auc_list = []

        # Loop over each seed folder
        for seed in seeds:
            # Build path to the single CSV file
            csv_path = os.path.join(
                base_conv_dir,
                ds,
                "csv",
                seed,
                pct_folder,
                f"results_{ds}_seed{seed}_p{pct_folder}.csv"
            )

            # If the file exists, read its AUC value
            if os.path.isfile(csv_path):
                df = pd.read_csv(csv_path, sep='\t|,', engine='python')
                # assume column is named exactly "AUC"
                auc_list.append(df.loc[0, "AUC"])
            else:
                print(f"Warning: missing file {csv_path}")

        # Compute the mean AUC for this percent (or NaN if none found)
        mean_auc = np.nan if not auc_list else float(np.mean(auc_list))

        # Append to summary
        summary.append({
            "Dataset": ds,
            "Percentage": percent_map[pct_folder],
            "Mean_AUC": mean_auc
        })

    # Convert to DataFrame and save
    summary_df = pd.DataFrame(summary)
    out_csv = os.path.join(
        base_mean_dir,
        ds,
        f"CNN_mean_metrics_{ds}_{percent_map['10']:.2f}p_to_{percent_map['75']:.2f}p.csv"
    )
    summary_df.to_csv(out_csv, index=False)
    print(f"Wrote summary for {ds} → {out_csv}")
```

```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches

# --- CONFIG ----------------------------------------------------------

datasets    = ["virtualshakespeare", "ml", "algo004", "comp"]
short_names = ["VS",               "ML",  "AL",     "CP"]
perc_folders= ["10", "25", "50", "75"]
percent_map = {"10": 0.10, "25": 0.25, "50": 0.50, "75": 0.75}

base_mean_dir = "drive/MyDrive/Colab Notebooks/GNN Research/mean_results"

# Base colors for each dataset
color_dict = {
    "virtualshakespeare": "#22CB1C",
    "ml":                 "#FFFF33",
    "algo004":            "#FF0000",
    "comp":               "#0080FF"
}

# Darken a color by a fixed factor
def darken(hex_color, amount=0.6):
    rgb = np.array(mcolors.to_rgb(hex_color))
    return mcolors.to_hex(rgb * amount)

# --- GATHER AUCs -----------------------------------------------------

# Structure: results[pct][model][ds] = mean_auc
results = { pct: {"GNN_iso":{}, "GNN_all":{}, "CNN":{}} for pct in perc_folders }

for pct_folder in perc_folders:
    p = percent_map[pct_folder]
    pct_str = f"{p:.2f}"

    # 1) CNN summary CSV
    #    path: mean_results/{ds}/CNN_mean_metrics_{ds}_0.10p_to_0.75p.csv
    #    columns: Dataset,Percentage,Mean_AUC,...
    cnn_dfs = {}
    for ds in datasets:
        cnn_csv = os.path.join(
            base_mean_dir, ds,
            f"CNN_mean_metrics_{ds}_0.10p_to_0.75p.csv"
        )
        cnn_dfs[ds] = pd.read_csv(cnn_csv, sep='\t|,', engine='python') \
                        .set_index("Percentage")

    # 2) GNN files for each ds
    for ds in datasets:
        gnn_csv = os.path.join(base_mean_dir, ds, f"mean_metrics_{ds}_{percent_map[pct_folder]}p.csv")
        df = pd.read_csv(gnn_csv, sep='\t|,', engine='python')

        # isolated
        iso = df.loc[df["Combination"] == ds, "Mean_AUC"]
        results[pct_folder]["GNN_iso"][ds] = float(iso.iloc[0]) if not iso.empty else np.nan

        # combined (all four)
        mask_all = df["Combination"].apply(
            lambda c: set(c.split("_")).issuperset(set(datasets))
        )
        allc = df.loc[mask_all, "Mean_AUC"]
        results[pct_folder]["GNN_all"][ds] = float(allc.iloc[0]) if not allc.empty else np.nan

        # CNN
        cnn_df = cnn_dfs[ds]
        results[pct_folder]["CNN"][ds] = float(cnn_df.loc[p, "Mean_AUC"])

# --- PLOT ------------------------------------------------------------

for pct_folder in perc_folders:
    p = percent_map[pct_folder]
    pct_label = int((1 - p) * 100)

    # Assemble DataFrame
    rows = []
    for model in ["GNN_iso", "GNN_all", "CNN"]:
        for ds in datasets:
            rows.append({
                "Dataset": ds,
                "Model": model,
                "Mean_AUC": results[pct_folder][model][ds]
            })
    plot_df = pd.DataFrame(rows)

    # Bar positions
    x     = np.arange(len(datasets))
    width = 0.25

    fig, ax = plt.subplots(figsize=(8, 4))

    for i, model in enumerate(["GNN_iso", "GNN_all", "CNN"]):
        offset = (i - 1) * width
        sub    = plot_df[plot_df["Model"] == model]
        base_colors = [color_dict[ds] for ds in sub["Dataset"]]

        if model == "GNN_iso":
            colors = base_colors
            hatch  = None
        elif model == "GNN_all":
            colors = [darken(c) for c in base_colors]
            hatch  = None
        else:  # CNN
            colors = [darken(c) for c in base_colors]
            hatch  = "///"

        bars = ax.bar(
            x + offset,
            sub["Mean_AUC"],
            width,
            label=model.replace("_"," "),
            color=colors,
            edgecolor="black",
            hatch=hatch
        )

        # annotate
        for bar in bars:
            h = bar.get_height()
            ax.text(
                bar.get_x() + bar.get_width()/2,
                h + 0.01,
                f"{h:.2f}",
                ha='center', va='bottom', fontsize=9
            )

    # Formatting
    ax.set_xticks(x)
    ax.set_xticklabels(short_names, fontsize=12)
    ax.set_xlabel("Dataset", fontsize=12)
    ax.set_ylabel("AUC", fontsize=12)
    ax.set_title(f"Model Comparison at {pct_label}% Progress", fontsize=14, fontweight='bold')
    ax.set_ylim(0, 1.05)
    # 1) Create custom legend patches with your new colors
    iso_handle = mpatches.Patch(
        facecolor="#fc03d7",
        edgecolor="black",
        label="GNN iso"
    )
    all_handle = mpatches.Patch(
        facecolor="#780166",
        edgecolor="black",
        label="GNN all"
    )
    cnn_handle = mpatches.Patch(
        facecolor="#780166",
        edgecolor="black",
        hatch="///",
        label="CNN"
    )

    # 2) Place them in a single‐column legend on the right, and enlarge the swatches
    ax.legend(
        handles=[iso_handle, all_handle, cnn_handle],
        title="Model",
        fontsize=10,
        title_fontsize=11,
        loc="center left",
        bbox_to_anchor=(1.02, 0.5),
        borderaxespad=0,
        ncol=1,
        handlelength=2,    # make the little color boxes wider
        handleheight=2     # make them taller, too
    )

    # 3) Shrink the plot area to make room for the legend
    plt.tight_layout(rect=[0, 0, 0.80, 1])

    plt.tight_layout(rect=[0, 0, 0.85, 1])
    out_path = os.path.join(base_mean_dir, f"compare_{pct_folder}p.png")
    #plt.savefig(out_path, dpi=600)
    plt.show()
    plt.close()
```

```
import os
import numpy as np
import pandas as pd

# Your folders
base = '/content/drive/MyDrive/Colab Notebooks/GNN Research/Convolutional NN'
datasets   = ['algo004', 'comp', 'ml', 'virtualshakespeare']
seeds      = [18, 61, 53, 29, 69, 42, 2, 21, 78, 99]
percentages = [10, 25, 50, 75]   # these are the pct_int values you used

# Prepare an accumulator
rows = []

for ds in datasets:
    for pct in percentages:
        precisions = []
        for seed in seeds:
            cm_path = os.path.join(
                base, ds, 'cm_npy',
                str(seed), str(pct),
                f'cm_{ds}_seed{seed}_p{pct}.npy'
            )
            if not os.path.exists(cm_path):
                print(f"  missing {cm_path}, skipping")
                continue

            tn, fp, fn, tp = np.load(cm_path).ravel()
            prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
            precisions.append(prec)

        if precisions:
            mean_prec = float(np.mean(precisions))
            std_prec  = float(np.std(precisions, ddof=1))
        else:
            mean_prec = np.nan
            std_prec  = np.nan

        rows.append({
            'Dataset':       ds,
            'Percentage':    pct / 100.0,
            'Mean_Precision': mean_prec,
            'Std_Precision':  std_prec,
            'N_Seeds':        len(precisions)
        })

# Build DataFrame and save
df = pd.DataFrame(rows)
out_path = os.path.join(base, 'mean_precision_by_pct.csv')
df.to_csv(out_path, index=False)
print(f"Saved mean precisions to {out_path}")
```

```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# --- CONFIG ----------------------------------------------------------
datasets    = ["virtualshakespeare", "ml", "algo004", "comp"]
short_names = ["VS",               "ML",  "AL",     "CP"]
perc_folders= ["10", "25", "50", "75"]
percent_map = {"10": 0.10, "25": 0.25, "50": 0.50, "75": 0.75}
base_mean_dir = "drive/MyDrive/Colab Notebooks/GNN Research/mean_results"

# Base (light) colors for each dataset
color_dict = {
    "virtualshakespeare": "#22CB1C",
    "ml":                 "#FFFF33",
    "algo004":            "#FF0000",
    "comp":               "#0080FF"
}

# --- PLOT GNN PRECISION ---------------------------------------------
for pct_folder in perc_folders:
    p = percent_map[pct_folder]
    label_pct = int((1-p)*100)

    # Read each dataset's GNN CSV and extract isolated & combined precision
    iso_vals = []
    all_vals = []
    for ds in datasets:
        path = os.path.join(base_mean_dir, ds, f"mean_metrics_{ds}_{percent_map[pct_folder]}p.csv")
        df = pd.read_csv(path, sep="\t|,", engine="python")

        # isolated
        iso = df.loc[df["Combination"] == ds, "Mean_Precision"]
        iso_vals.append(iso.iloc[0] if not iso.empty else np.nan)

        # combined (all four)
        mask = df["Combination"].apply(lambda c: set(c.split("_")).issuperset(set(datasets)))
        allp = df.loc[mask, "Mean_Precision"]
        all_vals.append(allp.iloc[0] if not allp.empty else np.nan)

    # Setup plot
    x     = np.arange(len(datasets))
    width = 0.35

    fig, ax = plt.subplots(figsize=(8,4))
    # Isolated bars
    bars_iso = ax.bar(
        x - width/2, iso_vals, width,
        label="GNN iso",
        color=[color_dict[ds] for ds in datasets],
        edgecolor="black",
        hatch=None
    )
    # Combined bars
    bars_all = ax.bar(
        x + width/2, all_vals, width,
        label="GNN all",
        color=[color_dict[ds] for ds in datasets],
        edgecolor="black",
        hatch="..."
    )

    # Annotate
    for bars in (bars_iso, bars_all):
        for bar in bars:
            h = bar.get_height()
            ax.text(
                bar.get_x() + bar.get_width()/2,
                h + 0.005,
                f"{h:.3f}",
                ha='center', va='bottom', fontsize=9
            )

    # Formatting
    ax.set_xticks(x)
    ax.set_xticklabels(short_names, fontsize=12)
    ax.set_xlabel("Dataset", fontsize=12)
    ax.set_ylabel("Mean Precision", fontsize=12)
    ax.set_title(f"GNN Precision: Isolated vs Combined at {label_pct}% Progress",
                 fontsize=14, fontweight='bold')
    ax.set_ylim(0, 0.3)

    # Legend
    iso_handle = mpatches.Patch(facecolor='white', edgecolor='black', label='GNN iso')
    all_handle = mpatches.Patch(facecolor='white', edgecolor='black', hatch='...', label='GNN all')
    ax.legend(handles=[iso_handle, all_handle],
              title="Model", fontsize=10, title_fontsize=11)

    plt.tight_layout()
    out = os.path.join(base_mean_dir, f"gnn_precision_{pct_folder}p.png")

    #plt.savefig(out, dpi=600, bbox_inches="tight")
    plt.show()
    plt.close()
```
